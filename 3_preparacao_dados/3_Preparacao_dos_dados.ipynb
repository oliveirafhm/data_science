## 3.10 - Consultas complexas

```
q = """
SELECT district, SUM (aadhaar_generated) 
FROM aadhaar_data GROUP BY district;
"""

pandasql.sqldf(q, locals()).head()
```

```
q = """
SELECT district, sub_district, SUM (aadhaar_generated) 
FROM aadhaar_data GROUP BY district, sub_district;
"""

pandasql.sqldf(q, locals()).head()
```
